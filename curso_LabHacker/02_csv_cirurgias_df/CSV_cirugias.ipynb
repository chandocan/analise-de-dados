```python
# importando as biblíotecas
import pandas as pd
```

```python
# informa ao matplotlib sobre grafico inline
%matplotlib inline 
```

```python
#  Vamos inporta apenas algumas colunas do arquivo
# vamos  importa as colunas 3 - hospital, 6 - Município , 7 - Complexidade, 8 - Carater Atendimento
# 12 sub Grupo procedimento, 14 - Procedimento
# Carregar o arquivo csv 
df = pd.read_csv('sih-janeiro-2017-cirurgias-eletiva-e-emergencia.csv', sep= ';', encoding= 'cp1252', usecols=[3,6,7,8,12,14])
# Troca o nome das colunas
df.columns = ['Hospital', 'Municipio', 'Complexidade', 'Carater Atendimento', 
'Sub Grupo Procedimento', 'Procedimento']

# exibir as primeiras linhas do dataframe
df.head(3) # por padrão mostra 5 mas nesse caso eu determinei 3
```

```python
# Descrever colunas númericas
df.describe()
```

```python
# lista os hospitais presentes nos dados
df['Hospital'].unique()
```

```python
# Quantas cirugias foram realizadas em cada hospital
df['Hospital'].value_counts()
```

```python
# quantas cirugias por sub grupo procedimento
df['Sub Grupo Procedimento'].value_counts()
```

```python
# plotar o gráfico por sub grupo procedimento
df['Sub Grupo Procedimento'].value_counts().plot.bar()
```

```python
# vamos criar um subconjunto dos dados originaís
df_hospbase = df[df['Hospital'] == '0010456 HBDF HOSPITAL DE BASE DO DISTRITO FEDERAL']
# linhas iniciais
df_hospbase.head(2)
```

```python
# para ver as linhas finais do dataframe
df_hospbase.tail(4)
```

```python
# fazer df mostra linhas aleatorias 
df_hospbase.sample(5)
```

```python
# posso ver a quantidade de procedimento realizados no hospital de base
df_hospbase['Procedimento'].value_counts()
```

```python
# procedimentos só relaciionado a amputação
# podemos subdividir o dataframe com base em pedaço de palavras
df_hospbase[df_hospbase['Procedimento'].str.contains('AMPUTA')].count()
```

```python
# dividir o dataframe original apenas pelo procedimento parto cesaríano
df_parto_cesariano = df[df['Procedimento'] == 'PARTO CESARIANO']
# primeiras linhas
df_parto_cesariano.head()
```

```python
# quantidade de partos cesariano por hospital
df_parto_cesariano['Hospital'].value_counts()
```

```python
# verificar a quantidade de partos cesarianos atendimento
df_parto_cesariano['Carater Atendimento'].value_counts()
```

```python
# vamos plotar um grafico de barras horizontal
df_parto_cesariano['Hospital'].value_counts().plot.barh()
```

```python
# vamos inverter a ordem é colocar um titulo para o grafico
# usar ascending=True
df_parto_cesariano['Hospital'].value_counts(ascending=True).plot.barh(title='Quantidade de partos cesariano por hospital')
```
